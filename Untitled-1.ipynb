```python
from tensorflow.keras.datasets import fashion_mnist

(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

print("x_train:", x_train.shape)
print("y_train:", y_train.shape)
print("x_test:", x_test.shape)
print("y_test:", y_test.shape)
```

## 2. ExploraciГіn inicial del dataset

En esta secciГіn se carga el conjunto de datos Fashion-MNIST y se realiza una exploraciГіn visual preliminar.  
Se identifican las clases disponibles, la forma de los datos y se visualizan ejemplos de imГЎgenes del conjunto de entrenamiento, con el fin de comprender la estructura del problema de clasificaciГіn.

```python
class_names = [
    "T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
    "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"
]

print("Primer label numГ©rico:", y_train[0])
print("Primera clase:", class_names[y_train[0]])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(3,3))
plt.imshow(x_train[0], cmap="gray")
plt.title(f"Clase: {class_names[y_train[0]]}")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(12, 8))

for i in range(15):
    plt.subplot(3, 5, i + 1)
    plt.imshow(x_train[i], cmap="gray")
    plt.title(class_names[y_train[i]])
    plt.axis("off")

plt.tight_layout()
plt.show()
```

## 3. Preprocesamiento de los datos

Antes de entrenar los modelos, es necesario preprocesar las imГЎgenes para que puedan ser utilizadas correctamente por las redes neuronales.  
En esta etapa se normalizan los valores de los pГӯxeles y se ajusta la forma de los datos segГәn el tipo de arquitectura a implementar.

```python
# NormalizaciГіn de imГЎgenes (de 0-255 a 0-1)
x_train_norm = x_train / 255.0
x_test_norm = x_test / 255.0

print("Valor mГӯnimo en x_train_norm:", x_train_norm.min())
print("Valor mГЎximo en x_train_norm:", x_train_norm.max())
```

```python
print("Forma original de x_train:", x_train.shape)
print("Forma normalizada de x_train_norm:", x_train_norm.shape)
```

### 3.1 PreparaciГіn de datos para red neuronal densa

Las redes neuronales densas no trabajan directamente con imГЎgenes bidimensionales, por lo que es necesario transformar cada imagen de 28x28 pГӯxeles en un vector unidimensional de 784 valores.  
Este proceso permite utilizar las imГЎgenes como entrada para una arquitectura totalmente conectada.

```python
# Aplanar imГЎgenes para red densa
x_train_flat = x_train_norm.reshape(60000, 28 * 28)
x_test_flat = x_test_norm.reshape(10000, 28 * 28)

print("Forma de x_train_flat:", x_train_flat.shape)
print("Forma de x_test_flat:", x_test_flat.shape)
```

## 4. Modelo 1: Red neuronal densa

En esta secciГіn se construye un primer modelo de clasificaciГіn utilizando una red neuronal densa (Artificial Neural Network, ANN).  
Este modelo servirГЎ como lГӯnea base para evaluar el desempeГұo inicial del problema de clasificaciГіn de imГЎgenes, antes de implementar una arquitectura convolutiva mГЎs especializada.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# Modelo denso
model_dense = Sequential([
    Dense(128, activation='relu', input_shape=(784,)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(10, activation='softmax')
])

model_dense.summary()
```

### 4.1 CompilaciГіn del modelo

Una vez definida la arquitectura, se compila el modelo especificando la funciГіn de pГ©rdida, el optimizador y la mГ©trica de evaluaciГіn.  
Dado que se trata de un problema de clasificaciГіn multiclase, se utiliza la funciГіn de pГ©rdida `sparse_categorical_crossentropy`, junto con el optimizador Adam y la mГ©trica de exactitud.

```python
model_dense.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("Modelo compilado correctamente.")
```

### 4.2 Entrenamiento del modelo

En esta etapa se entrena la red neuronal densa utilizando los datos preprocesados.  
Se emplea un subconjunto de validaciГіn para monitorear el comportamiento del modelo durante el entrenamiento y observar su capacidad de generalizaciГіn.

```python
history_dense = model_dense.fit(
    x_train_flat,
    y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)
```

### 4.3 EvaluaciГіn del modelo

Una vez finalizado el entrenamiento, se evalГәa el modelo utilizando el conjunto de prueba.  
Esto permite medir el rendimiento real del modelo frente a datos no vistos previamente.

```python
test_loss, test_acc = model_dense.evaluate(x_test_flat, y_test)

print("Test accuracy:", test_acc)
print("Test loss:", test_loss)
```

### 4.4 VisualizaciГіn del entrenamiento

Para analizar el comportamiento del modelo durante el entrenamiento, se grafican las mГ©tricas de exactitud y pГ©rdida tanto para entrenamiento como para validaciГіn.  
Esto permite observar la evoluciГіn del aprendizaje y detectar posibles problemas como sobreajuste o subajuste.

```python
plt.figure(figsize=(8,5))
plt.plot(history_dense.history['accuracy'], label='Entrenamiento')
plt.plot(history_dense.history['val_accuracy'], label='ValidaciГіn')
plt.title('Accuracy del modelo denso')
plt.xlabel('Гүpocas')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(history_dense.history['loss'], label='Entrenamiento')
plt.plot(history_dense.history['val_loss'], label='ValidaciГіn')
plt.title('Loss del modelo denso')
plt.xlabel('Гүpocas')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

### 4.5 Predicciones del modelo denso

Con el fin de observar el comportamiento del modelo sobre imГЎgenes individuales, se generan predicciones sobre ejemplos del conjunto de prueba.  
Esto permite comparar visualmente la clase real de la imagen con la clase predicha por la red neuronal.

```python
predictions_dense = model_dense.predict(x_test_flat)
print("Forma de predictions_dense:", predictions_dense.shape)
```

```python
import numpy as np

plt.figure(figsize=(12, 8))

for i in range(12):
    plt.subplot(3, 4, i + 1)
    plt.imshow(x_test[i], cmap="gray")
    
    pred_class = np.argmax(predictions_dense[i])
    real_class = y_test[i]
    
    plt.title(f"Real: {class_names[real_class]}\nPred: {class_names[pred_class]}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

## 5. Modelo 2: Red neuronal convolutiva

En esta secciГіn se implementa una red neuronal convolutiva (CNN) para abordar el mismo problema de clasificaciГіn de imГЎgenes.  
A diferencia del modelo denso, esta arquitectura permite aprovechar la estructura espacial de las imГЎgenes, lo que generalmente mejora el rendimiento en tareas de visiГіn por computador.

```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten

model_cnn = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),
    
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(10, activation='softmax')
])

model_cnn.summary()
```

### 5.2 CompilaciГіn del modelo convolutivo

Una vez definida la arquitectura convolutiva, se compila el modelo utilizando una funciГіn de pГ©rdida adecuada para clasificaciГіn multiclase, el optimizador Adam y la mГ©trica de exactitud.

```python
model_cnn.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("Modelo CNN compilado correctamente.")
```

### 5.3 Entrenamiento del modelo convolutivo

En esta etapa se entrena la red neuronal convolutiva utilizando las imГЎgenes preprocesadas con su estructura espacial original.  
Se incorpora un conjunto de validaciГіn para monitorear el desempeГұo del modelo durante el entrenamiento.

```python
# Ajustar dimensiГіn para CNN: (n, 28, 28, 1)
x_train_cnn = x_train_norm.reshape(60000, 28, 28, 1)
x_test_cnn = x_test_norm.reshape(10000, 28, 28, 1)

print("Forma de x_train_cnn:", x_train_cnn.shape)
print("Forma de x_test_cnn:", x_test_cnn.shape)
```

```python
history_cnn = model_cnn.fit(
    x_train_cnn,
    y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)
```

### 5.4 EvaluaciГіn del modelo convolutivo

Una vez finalizado el entrenamiento, se evalГәa el modelo convolutivo sobre el conjunto de prueba para medir su capacidad de generalizaciГіn frente a imГЎgenes no vistas.

```python
test_loss_cnn, test_acc_cnn = model_cnn.evaluate(x_test_cnn, y_test)

print("Test accuracy CNN:", test_acc_cnn)
print("Test loss CNN:", test_loss_cnn)
```

### 5.5 VisualizaciГіn del entrenamiento del modelo convolutivo

Se grafican las mГ©tricas de exactitud y pГ©rdida del modelo convolutivo para analizar su evoluciГіn durante el entrenamiento y comparar su comportamiento con el modelo denso.

```python
plt.figure(figsize=(8,5))
plt.plot(history_cnn.history['accuracy'], label='Entrenamiento')
plt.plot(history_cnn.history['val_accuracy'], label='ValidaciГіn')
plt.title('Accuracy del modelo CNN')
plt.xlabel('Гүpocas')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(history_cnn.history['loss'], label='Entrenamiento')
plt.plot(history_cnn.history['val_loss'], label='ValidaciГіn')
plt.title('Loss del modelo CNN')
plt.xlabel('Гүpocas')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

### 5.6 Predicciones del modelo convolutivo

Se generan predicciones sobre imГЎgenes del conjunto de prueba para observar visualmente el desempeГұo del modelo y comparar la clase real con la clase predicha.

```python
predictions_cnn = model_cnn.predict(x_test_cnn)
print("Forma de predictions_cnn:", predictions_cnn.shape)
```

```python
plt.figure(figsize=(12, 8))

for i in range(12):
    plt.subplot(3, 4, i + 1)
    plt.imshow(x_test[i], cmap="gray")
    
    pred_class = np.argmax(predictions_cnn[i])
    real_class = y_test[i]
    
    plt.title(f"Real: {class_names[real_class]}\nPred: {class_names[pred_class]}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
